<h3 id="for">驴Qu茅 es el bucle <code>for</code>?</h3>

El bucle <code>for</code> permite ejecutar un bloque de c贸digo m煤ltiples veces. Por ejemplo, podr铆as usarlo si quisieras imprimir cada elemento en una lista. Usemos el bucle <code>for</code> para imprimir todos los a帽os presentes en la lista <code>dates</code>:

Esto puede hacerse de la siguiente forma:

```python
# Ejemplo de un bucle for

dates = [1982,1980,1973]
N = len(dates)

for i in range(N):
    print(dates[i])     
```

En Python podemos acceder directamente a los elementos de la lista de la siguiente forma:

```python
# Ejemplo de un bucle for en una lista

for year in dates:  
    print(year)   
```

Podemos modificar los elementos en una lista:

```python
# Usar el bucle for para cambiar los elementos de la lista

squares = ['red', 'yellow', 'green', 'purple', 'blue']

for i in range(0, 5):
    print("Before square ", i, 'is',  squares[i])
    squares[i] = 'weight'
    print("After square ", i, 'is',  squares[i])
```

Podemos acceder al 铆ndice y los elementos de una lista de la siguiente manera:

```python
# Bucle for en una lista haciendo iteraciones con el valor del 铆ndice y el elemento de la lista

squares=['red', 'yellow', 'green', 'purple', 'blue']

for i, square in enumerate(squares):
    print(i, square)
```

<h3 id="while">驴Qu茅 es el bucle <code>while</code>?</h3>

Como puedes observar, el bucle <code>for</code> es utilizado para controlar el flujo de la repetici贸n. Sin embargo, 驴Qu茅 pasa cuando no sabemos en que momento detener el bucle? 驴Y si queremos mantener la ejecuci贸n de un bloque de c贸digo hasta que se cumpla cierta condici贸n? El bucle <code>while</code> es una herramienta para ejecutar repetidamente en base a una condici贸n. El bloque de c贸digo seguir谩 ejecut谩ndose hasta que la condici贸n l贸gica dada regresa un valor **Falso**.

Supongamos que deseamos interactuar sobre la lista <code>dates</code> y detenernos en el a帽o 1973, despu茅s imprimir el numero de iteraciones. Esto puede realizarse con el siguiente bloque de c贸digo:

```python
# Ejemplo de Bucle While

dates = [1982, 1980, 1973, 2000]

i = 0
year = 0

while(year != 1973):
    year = dates[i]
    i = i + 1
    print(year)

print("It took ", i ,"repetitions to get out of loop.")
```
